```python
import pandas as pd
```

```python
content=[]
index=0
for i in range(len(full_list)):# 1 mas el numero de frames len(full_list)
    a=pd.read_csv(full_list[index], sep=" ", index_col=None,header=None)
    a.columns = ["class","xtop","ytop","alto","ancho", "proba"]
    b=full_list[index][94:147]
    patron = re.compile(r'_([0-9]+)\.txt')
    match = patron.search(b)
    nu = match.group(1)
    a["frame"]=nu
    content.append(a)
    data_frame = pd.concat(content)
    index +=1
```

```python
content
```

```python
n = data_frame["frame"].str.strip(".t" )
na = n.str.strip("." )
data_frame["frames"]=na
df=data_frame.drop(columns = ["frame"])
df['frames'] = pd.to_numeric(df['frames'])
frames=df.sort_values(by="frames")
frames
```

```python
frames.to_excel("/content/drive/MyDrive/ANEXOS_TESIS/6.Coordenadas_mapa_calor/framesv/frames_nacho2.xlsx")
```

```python
sorted_df = frames.sort_values(by="proba", ascending=False).drop_duplicates(subset="frames")

result_df = sorted_df.groupby("frames").first().reset_index()
```
