```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN

# Generar datos de ejemplo
np.random.seed(0)
X = np.random.randn(21, 2)

# Agregar algunos puntos atípicos
outliers = np.array([[10, 10], [10, 12], [12, 10]])
X = np.vstack([X, outliers])

# Aplicar DBSCAN
dbscan = DBSCAN(eps=1, min_samples=3)
dbscan.fit(X)

# Obtener etiquetas de cluster y puntos atípicos
labels = dbscan.labels_
core_samples_mask = np.zeros_like(labels, dtype=bool)
core_samples_mask[dbscan.core_sample_indices_] = True

# Número de clusters encontrados y puntos atípicos
n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
n_noise_ = list(labels).count(-1)

# Crear un gráfico
unique_labels = set(labels)
colors = [plt.cm.Spectral(each) for each in np.linspace(0, 1, len(unique_labels))]

plt.figure(figsize=(8, 6))

for k, col in zip(unique_labels, colors):
    if k == -1:
        # Puntos atípicos en negro
        col = [0, 0, 0, 1]

    class_member_mask = (labels == k)

    xy = X[class_member_mask & core_samples_mask]
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=tuple(col), markeredgecolor='k', markersize=14)

    xy = X[class_member_mask & ~core_samples_mask]
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=tuple(col), markeredgecolor='k', markersize=6)

plt.title(f'Número estimado de clusters: {n_clusters_}')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN

# Generar datos de ejemplo
np.random.seed(0)
X = np.random.randn(21, 2)

# Agregar algunos puntos atípicos
outliers = np.array([[7, 7]])
X = np.vstack([X, outliers])

# Aplicar DBSCAN
dbscan = DBSCAN(eps=1, min_samples=3)
dbscan.fit(X)

# Obtener etiquetas de cluster
labels = dbscan.labels_

# Calcular el tamaño de cada cluster
unique_labels, cluster_sizes = np.unique(labels, return_counts=True)

# Encontrar el índice del cluster más grande
largest_cluster_index = cluster_sizes.argmax()

# Encontrar los puntos que pertenecen al cluster más grande
largest_cluster_points = X[labels == unique_labels[largest_cluster_index]]

# Calcular el centroide del cluster más grande
centroid = np.mean(largest_cluster_points, axis=0)

# Crear un gráfico
colors = [plt.cm.Spectral(each) for each in np.linspace(0, 1, len(unique_labels))]

plt.figure(figsize=(8, 6))

for k, col in zip(unique_labels, colors):
    if k == -1:
        col = [0, 0, 0, 1]

    class_member_mask = (labels == k)

    xy = X[class_member_mask]
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=tuple(col), markeredgecolor='k', markersize=6)

# Dibujar el centroide del cluster más grande
plt.plot(centroid[0], centroid[1], 'o', markerfacecolor='red', markeredgecolor='k', markersize=10)
plt.title(f'Punto central de la mayoría de los puntos')
plt.show()

# Imprimir las coordenadas del punto central
print(f'Punto central: ({centroid[0]:.2f}, {centroid[1]:.2f})')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from scipy.spatial.distance import pdist, squareform

# Puntos
points = np.array([[633, 267],[700,300], [700, 240], [900, 480]])

# Calcular las distancias entre todos los pares de puntos
distances = pdist(points)

# Convertir las distancias en una matriz cuadrada
dist_matrix = squareform(distances)

# Calcular la distancia mínima entre los puntos
min_distance = np.min(dist_matrix[np.nonzero(dist_matrix)])

# Calcular el punto central
central_point = np.mean(points, axis=0)

# Aplicar DBSCAN considerando la distancia mínima como "eps"
dbscan = DBSCAN(eps=min_distance, min_samples=2)
dbscan.fit(points)

# Etiquetas de cluster
labels = dbscan.labels_

# Obtener los puntos pertenecientes al cluster con etiqueta 0
cluster_points = points[labels == 0]

# Calcular el punto central del cluster
cluster_center = np.mean(cluster_points, axis=0)

# Crear un gráfico
plt.figure(figsize=(8, 6))

# Dibujar los puntos originales
plt.scatter(points[:, 0], points[:, 1], c='blue', label='Puntos')

# Dibujar el punto central del cluster
plt.scatter(cluster_center[0], cluster_center[1], c='red', marker='x', s=100, label='Punto Central')

# Etiquetas y leyenda
plt.xlabel('Coordenada X')
plt.ylabel('Coordenada Y')
plt.legend()

# Mostrar el gráfico
plt.grid()
plt.title('Puntos y Punto Central')
plt.show()
```
